```python
import pandas as pd
import os

# Go to project root
os.chdir('..')
print("Working directory:", os.getcwd())
```

```python
data_dir = 'data/raw_data/'
files = sorted(os.listdir(data_dir))

datasets = {}

for file in files:
    if file.endswith('.csv'):
        df = pd.read_csv(os.path.join(data_dir, file))
        name = file.replace('.csv', '')
        datasets[name] = df
        print(f"\n{'='*50}")
        print(f"FILE: {file}")
        print(f"Shape: {df.shape}")
        print(f"\nColumn types:\n{df.dtypes}")
        print(f"\nFirst 3 rows:\n{df.head(3)}")
```

```python
print("DATA QUALITY SUMMARY")
print("="*50)
for file in files:
    if file.endswith('.csv'):
        df = pd.read_csv(os.path.join(data_dir, file))
        nulls = df.isnull().sum().sum()
        dupes = df.duplicated().sum()
        print(f"{file}")
        print(f"  → Nulls: {nulls}, Duplicates: {dupes}")
```

```python
date_cols = {
    '01_fund_master':           'launch_date',
    '02_nav_history':           'date',
    '03_aum_by_fund_house':     'date',
    '04_monthly_sip_inflows':   'month',
    '05_category_inflows':      'month',
    '06_industry_folio_count':  'month',
    '08_investor_transactions':  'transaction_date',
    '09_portfolio_holdings':     'portfolio_date',
    '10_benchmark_indices':      'date'
}

datasets = {}

for file in sorted(os.listdir(data_dir)):
    if file.endswith('.csv'):
        name = file.replace('.csv', '')
        df = pd.read_csv(os.path.join(data_dir, file))
        if name in date_cols:
            df[date_cols[name]] = pd.to_datetime(df[date_cols[name]])
        datasets[name] = df

print("✅ All 10 files loaded with correct date types")
```

```python
master_codes = set(datasets['01_fund_master']['amfi_code'].unique())
nav_codes    = set(datasets['02_nav_history']['amfi_code'].unique())
perf_codes   = set(datasets['07_scheme_performance']['amfi_code'].unique())
txn_codes    = set(datasets['08_investor_transactions']['amfi_code'].unique())
port_codes   = set(datasets['09_portfolio_holdings']['amfi_code'].unique())

print("=== AMFI CODE VALIDATION ===")
print(f"fund_master           : {len(master_codes)} codes")
print(f"nav_history           : {len(nav_codes)} codes")
print(f"scheme_performance    : {len(perf_codes)} codes")
print(f"investor_transactions : {len(txn_codes)} codes")
print(f"portfolio_holdings    : {len(port_codes)} codes")

missing = master_codes - nav_codes
print(f"\nCodes in master but missing from nav_history: {missing if missing else '✅ None'}")
```
